```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set up the webdriver (make sure the path is correct)
driver = webdriver.Chrome()  # or webdriver.Firefox(), etc.

# Navigate to the Wikipedia page about web scraping
driver.get("https://en.wikipedia.org/wiki/Web_scraping")

# Wait for the link to be clickable and then click it
link = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//a[@href='/wiki/Archive.today']"))
)

# Click on the link
link.click()

# Optional: Wait for a bit to see the result
time.sleep(5)

# Close the driver
driver.quit()
```
