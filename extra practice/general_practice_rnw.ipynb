```python
from decimal import Decimal, ROUND_HALF_UP

# Input values to test
values = [Decimal('0.86'), Decimal('0.87'), Decimal('0.88'), Decimal('0.89'),
          Decimal('0.90'), Decimal('0.91'), Decimal('0.92'), Decimal('0.93'),
          Decimal('0.94'), Decimal('0.95'), Decimal('0.96'), Decimal('0.97')]

# Process each value
for value in values:
    # Check if the value is an exact multiple of 0.05
    if value % Decimal('0.05') == 0:
        rounded_value = value
    else:
        # Round to the nearest multiple of 0.10
        rounded_value = (value + Decimal('0.05')).quantize(Decimal('0.10'), rounding=ROUND_HALF_UP)

    # Output the original and rounded values
    print(f"{value} rounds to {rounded_value}")
```
